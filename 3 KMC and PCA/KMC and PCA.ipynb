# K-means Clustering

In this this exercise, you will implement the K-means algorithm and use it for image compression. You will first start on an example 2D dataset that will help you gain an intuition of how the K-means algorithm works.

```python
import math
import numpy as np
import pandas as pd
from datetime import datetime

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
plt.style.use('seaborn-whitegrid')

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report,confusion_matrix
```

```python
import scipy.io
```

```python
data = scipy.io.loadmat('ex7data2.mat')
data = pd.DataFrame(data['X'])
data.columns = ['x1','x2']
plt.plot(data['x1'],data['x2'],'.')
```

```python
kmeans = KMeans(n_clusters=3,random_state=101)
kmeans.fit(data)
kmeans.cluster_centers_
```

```python
plt.scatter(data['x1'], data['x2'], c=kmeans.labels_.astype(float), cmap=plt.cm.viridis)
```

# Principal Component Analysis

In this exercise, you will use principal component analysis (PCA) to perform dimensionality reduction.

```python
data = scipy.io.loadmat('ex7data1.mat')
data = pd.DataFrame(data['X'])
data.columns = ['x1','x2']
plt.plot(data['x1'],data['x2'],'.')
```

```python
scaler = StandardScaler()
scaler.fit(data)
scaled_data = scaler.transform(data)
scaled_data = pd.DataFrame(scaled_data)
scaled_data.columns = ['x1','x2']
plt.plot(scaled_data['x1'],scaled_data['x2'],'.')
```

```python
pca = PCA(n_components=2,random_state=101)
pca.fit(data)
x_pca = pca.transform(data)
print(pca.components_)
print(pca.explained_variance_)
```
